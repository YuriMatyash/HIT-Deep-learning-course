### 1. Sigmoid Function Gradient

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

1. Differentiate $\sigma(x)$ with respect to $x$.  
$$
\sigma(x) = \frac{1}{1 + e^{-x}} = (1 + e^{-x})^{-1}
$$  


$$
u = 1 + e^{-x} 
$$

$$
\sigma(u) = u^{-1}
$$

$$
\frac{d\sigma}{dx} = \frac{d\sigma}{du} \cdot \frac{du}{dx}  = ((-1) \cdot (1 + e^{-x})^{-2})\cdot((-1) \cdot {e}^{-x}) = \frac{-1}{(1 + e^{-x})^2} \cdot (-e^{-x}) = \frac{e^{-x}}{(1 + e^{-x})^2}
$$

2. Show that the derivative of $\sigma(x)$ can be expressed in terms of $\sigma(x)$ itself

$$
\frac{d\sigma}{dx} = \frac{e^{-x}}{(1 + e^{-x})^2} = \frac{e^{-x}}{(1 + e^{-x})} \cdot \frac{1}{(1 + e^{-x})} = \frac{e^{-x}}{(1 + e^{-x})} \cdot \sigma(x)
$$


### 2. Tanh Function Gradient
$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

1. Differentiate $\tanh(x)$ with respect to $x$.
$$
\tanh(x) =  2\sigma(2x) - 1 = 2 \cdot \frac{1}{1 + e^{-2x}}-1
$$

$$
\frac{d}{dx}\tanh(x) = 2 \cdot \frac{d}{dx}\frac{1}{1 + e^{-2x}} + 0= 2 \cdot \frac{d}{dx}\sigma(2x)
$$


$$
\frac{d\sigma(2x)}{dx} = \frac{d}{dx}\frac{1}{1 + e^{-2x}} = \frac{d}{dx}(1 + e^{-2x})^{-1} = -1\cdot(1+e^{-2x})^{-2}\cdot \frac{d}{dx}(1+e^{-2x})
$$

$$
\frac{d}{dx}(1+e^{-2x}) = 0 + e^{-2x}\cdot(-2) = -2e^{-2x}
$$

$$
\frac{d\sigma(2x)}{dx} = -1\cdot(1+e^{-2x})^{-2}\cdot -2e^{-2x} = \frac{2e^{-2x}}{(1+e^{-2x})^{2}}
$$

$$
\frac{d}{dx}\tanh(x) = 2\cdot \frac{2e^{-2x}}{(1+e^{-2x})^{2}} = \frac{4e^{-2x}}{(1+e^{-2x})^{2}}
$$


2. Show that the derivative of $\tanh(x)$ can be expressed as:

   $$
   \frac{d\tanh(x)}{dx} = 1 - \tanh(x)^2
   $$


$$
1 - \tanh(x)^2 = 1 - (\frac{e^x - e^{-x}}{e^x + e^{-x}})^{2}
$$

$$
= \frac{(e^x + e^{-x})^{2}}{(e^x + e^{-x})^2}-\frac{(e^x - e^{-x})^{2}}{(e^x + e^{-x})^2}
$$

$$
= \frac{(e^x + e^{-x})^{2} - (e^x - e^{-x})^{2}}{(e^x + e^{-x})^2}
$$

$$
= \frac{(e^{2x} + 2e^xe^{-x} + e^{-2x}) - (e^{2x} - 2e^xe^{-x} + e^{-2x})}{(e^x + e^{-x})^2}
$$

$$
= \frac{(e^{2x} + 2 + e^{-2x}) - (e^{2x} - 2 + e^{-2x})}{(e^x + e^{-x})^2}
$$

$$
= \frac{e^{2x} + 4 + e^{-2x} - e^{2x} -e^{-2x}}{(e^x + e^{-x})^2}
$$

$$
= \frac{4}{(e^x + e^{-x})^2}
$$

$$
= \frac{4}{(e^x + e^{-x})^2} \cdot \frac{e^{-2x}}{e^{-2x}} = \frac{4e^{-2x}}{(e^{x} + e^{-x})^2\cdot e^{-2x}}
$$

$$
= \frac{4e^{-2x}}{((e^{x} + e^{-x})\cdot e^{-x})^2}
$$

$$
= \frac{4e^{-2x}}{(e^{x}e^{-x} + e^{-2x})^2}
$$

$$
= \frac{4e^{-2x}}{(1 + e^{-2x})^2} = \frac{d}{dx}\tanh(x)
$$
